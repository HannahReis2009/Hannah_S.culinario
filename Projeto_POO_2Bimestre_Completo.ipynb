```python
# ==========================================================
# SISTEMA DE RECEITAS - PROJETO POO COM BANCO DE DADOS
# Desenvolvido por: Hannah Reis
# ==========================================================

# -----------------------------
# IMPORTA√á√ÉO DE BIBLIOTECAS
# -----------------------------
import mysql.connector  # Biblioteca para conectar ao MySQL ‚Äì Hannah Reis

# -----------------------------
# PARTE 1 ‚Äì CLASSE PRINCIPAL
# - Atributos
# - Construtor
# - M√©todo de exibi√ß√£o
# -----------------------------
class ReceitaHannahReis:
    def __init__(self, nome, ingredientes, tempo_preparo):  # Construtor ‚Äì Hannah Reis
        self.nome = nome  # Nome da receita ‚Äì Hannah Reis
        self.ingredientes = ingredientes  # Lista de ingredientes ‚Äì Hannah Reis
        self.__tempo_preparo = tempo_preparo  # Tempo privado ‚Äì Hannah Reis (Parte 3)

    def exibir_dados_hannah_reis(self):  # M√©todo para exibir dados da receita ‚Äì Hannah Reis
        print(f"\nReceita: {self.nome}")
        print("Ingredientes:")
        for ingrediente in self.ingredientes:
            print(f" - {ingrediente}")
        print(f"Tempo de preparo: {self.__tempo_preparo} minutos")
        print(f"Classifica√ß√£o: {self.classificar_preparo_hannah_reis()}")  # Chamada de m√©todo l√≥gico (Parte 3)

# -----------------------------
# PARTE 2 ‚Äì M√âTODOS DE A√á√ÉO
# -----------------------------
    def adicionar_ingrediente_hannah_reis(self, ingrediente):
        self.ingredientes.append(ingrediente)

# -----------------------------
# PARTE 3 ‚Äì ENCAPSULAMENTO
# - get/set para tempo_preparo
# - L√≥gica interna: classificar preparo
# -----------------------------
    def get_tempo_preparo_hannah_reis(self):
        return self.__tempo_preparo

    def set_tempo_preparo_hannah_reis(self, novo_tempo):
        if novo_tempo > 0:
            self.__tempo_preparo = novo_tempo
        else:
            print("Tempo inv√°lido.")

    def classificar_preparo_hannah_reis(self):  # M√©todo com l√≥gica interna ‚Äì Hannah Reis
        if self.__tempo_preparo <= 20:
            return "R√°pido"
        elif self.__tempo_preparo <= 60:
            return "M√©dio"
        else:
            return "Demorado"

# -----------------------------
# PARTE 2 ‚Äì SUBCLASSE COM HERAN√áA
# -----------------------------
class ReceitaSobremesaHannahReis(ReceitaHannahReis):
    def __init__(self, nome, ingredientes, tempo_preparo, tipo_doce):
        super().__init__(nome, ingredientes, tempo_preparo)
        self.tipo_doce = tipo_doce  # Novo atributo exclusivo

    def exibir_dados_hannah_reis(self):  # Sobrescrita do m√©todo (polimorfismo ‚Äì Parte 4)
        super().exibir_dados_hannah_reis()
        print(f"Tipo de doce: {self.tipo_doce}")

# -----------------------------
# PARTE 4 ‚Äì FUN√á√ïES DE BANCO DE DADOS (MySQL)
# -----------------------------

# Conectar ao banco MySQL
def conectar_banco_hannah_reis():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="receitas_hannah_reis"
    )

# Inserir nova receita no banco
def inserir_receita_hannah_reis(receita):
    conn = conectar_banco_hannah_reis()
    cursor = conn.cursor()
    sql = "INSERT INTO receitas (nome, ingredientes, tempo_preparo) VALUES (%s, %s, %s)"
    ingredientes_str = ", ".join(receita.ingredientes)
    dados = (receita.nome, ingredientes_str, receita.get_tempo_preparo_hannah_reis())
    cursor.execute(sql, dados)
    conn.commit()
    cursor.close()
    conn.close()
    print("‚úÖ Receita inserida com sucesso!")

# Listar todas as receitas do banco
def listar_receitas_hannah_reis():
    conn = conectar_banco_hannah_reis()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM receitas")
    receitas = cursor.fetchall()
    if not receitas:
        print("‚ö†Ô∏è Nenhuma receita cadastrada.")
    for (id, nome, ingredientes, tempo) in receitas:
        print(f"\nüî∏ ID: {id}\nüî∏ Nome: {nome}\nüî∏ Ingredientes: {ingredientes}\nüî∏ Tempo: {tempo} minutos")
    cursor.close()
    conn.close()

# Deletar receita pelo ID
def deletar_receita_hannah_reis(id_receita):
    conn = conectar_banco_hannah_reis()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM receitas WHERE id = %s", (id_receita,))
    conn.commit()
    cursor.close()
    conn.close()
    print("üóëÔ∏è Receita deletada com sucesso!")

# -----------------------------
# PARTE 5 ‚Äì MENU DE INTERA√á√ÉO
# -----------------------------
def menu_hannah_reis():
    while True:
        print("\nüçΩÔ∏è SISTEMA DE RECEITAS - HANNAH REIS")
        print("1 - Adicionar nova receita")
        print("2 - Listar receitas")
        print("3 - Deletar receita")
        print("0 - Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            nome = input("Digite o nome da receita: ")
            ingredientes = input("Informe os ingredientes (separados por v√≠rgula): ").split(",")
            tempo = int(input("Informe o tempo de preparo (em minutos): "))
            receita = ReceitaHannahReis(nome, [i.strip() for i in ingredientes], tempo)
            inserir_receita_hannah_reis(receita)

        elif opcao == "2":
            listar_receitas_hannah_reis()

        elif opcao == "3":
            try:
                id_receita = int(input("Digite o ID da receita que deseja deletar: "))
                deletar_receita_hannah_reis(id_receita)
            except ValueError:
                print("‚ö†Ô∏è ID inv√°lido.")

        elif opcao == "0":
            print("‚úÖ Sistema encerrado. At√© logo!")
            break
        else:
            print("‚ö†Ô∏è Op√ß√£o inv√°lida, tente novamente.")

# -----------------------------
# EXECUTAR SISTEMA
# -----------------------------
menu_hannah_reis()
```
