## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**

**C√≥digo:**

```python
import mysql.connector  # Hannah Reis

# Hannah Reis
class ReceitaHannahReis:  # Hannah Reis
    def __init__(self, nome, ingredientes, tempo_preparo):  # Hannah Reis
        self.nome = nome  # Hannah Reis
        self.ingredientes = ingredientes  # Hannah Reis
        self.__tempo_preparo = tempo_preparo  # Hannah Reis

    def adicionar_ingrediente_hannah_reis(self, ingrediente):  # Hannah Reis
        self.ingredientes.append(ingrediente)  # Hannah Reis
        print(f"Ingrediente '{ingrediente}' adicionado √† receita '{self.nome}'.")  # Hannah Reis

    def get_tempo_preparo_hannah_reis(self):  # Hannah Reis
        return self.__tempo_preparo  # Hannah Reis

    def set_tempo_preparo_hannah_reis(self, novo_tempo):  # Hannah Reis
        if novo_tempo > 0:  # Hannah Reis
            print(f"Tempo de preparo da receita '{self.nome}' alterado de {self.__tempo_preparo} para {novo_tempo} minutos.")  # Hannah Reis
            self.__tempo_preparo = novo_tempo  # Hannah Reis
        else:  # Hannah Reis
            print("Tempo inv√°lido. O valor deve ser maior que zero.")  # Hannah Reis

    def classificar_preparo_hannah_reis(self):  # Hannah Reis (l√≥gica interna)
        if self.__tempo_preparo <= 20:  # Hannah Reis
            return "R√°pido"  # Hannah Reis
        elif self.__tempo_preparo <= 60:  # Hannah Reis
            return "M√©dio"  # Hannah Reis
        else:  # Hannah Reis
            return "Demorado"  # Hannah Reis

    def exibir_dados_hannah_reis(self):  # Hannah Reis
        print(f"Receita: {self.nome}")  # Hannah Reis
        print("Ingredientes:")  # Hannah Reis
        for ingrediente in self.ingredientes:  # Hannah Reis
            print(f"- {ingrediente}")  # Hannah Reis
        print(f"Tempo de preparo: {self.__tempo_preparo} minutos")  # Hannah Reis
        print(f"Classifica√ß√£o: {self.classificar_preparo_hannah_reis()}")  # Hannah Reis
        print("-" * 40)  # Hannah Reis

# Subclasse herdando de ReceitaHannahReis  # Hannah Reis
class ReceitaSobremesaHannahReis(ReceitaHannahReis):  # Hannah Reis
    def __init__(self, nome, ingredientes, tempo_preparo, tipo_doce):  # Hannah Reis
        super().__init__(nome, ingredientes, tempo_preparo)  # Hannah Reis
        self.tipo_doce = tipo_doce  # Hannah Reis

    def exibir_dados_hannah_reis(self):  # Hannah Reis (polimorfismo)
        super().exibir_dados_hannah_reis()  # Hannah Reis (chama o m√©todo da classe base)
        print(f"Tipo de doce: {self.tipo_doce}")  # Hannah Reis
        print("=" * 40)  # Hannah Reis

# Hannah Reis
class BancoDeDadosHannahReis:  # Hannah Reis
    def __init__(self, host, user, password, nome_banco):  # Hannah Reis
        self.conexao = mysql.connector.connect(  # Hannah Reis (conecta ao banco de dados MySQL)
            host=host,  # Hannah Reis (endere√ßo do servidor MySQL)
            user=user,  # Hannah Reis (usu√°rio MySQL)
            password=password,  # Hannah Reis (senha do MySQL)
            database=nome_banco  # Hannah Reis (nome do banco de dados)
        )  # Hannah Reis
        self.cursor = self.conexao.cursor()  # Hannah Reis
        self.criar_tabela_hannah_reis()  # Hannah Reis

    def criar_tabela_hannah_reis(self):  # Hannah Reis
        self.cursor.execute('''  # Hannah Reis
            CREATE TABLE IF NOT EXISTS receitas (  # Hannah Reis
                id INT AUTO_INCREMENT PRIMARY KEY,  # Hannah Reis (ajustado para MySQL)
                nome VARCHAR(255),  # Hannah Reis
                ingredientes TEXT,  # Hannah Reis
                tempo_preparo INT,  # Hannah Reis
                tipo_doce VARCHAR(100)  # Hannah Reis (nova coluna para sobremesas)
            )  # Hannah Reis
        ''')  # Hannah Reis
        self.conexao.commit()  # Hannah Reis

    def inserir_receita_hannah_reis(self, receita):  # Hannah Reis
        ingredientes_str = ", ".join(receita.ingredientes)  # Hannah Reis
        tipo_doce = getattr(receita, 'tipo_doce', None)  # Hannah Reis (verifica se o atributo 'tipo_doce' existe)
        self.cursor.execute("INSERT INTO receitas (nome, ingredientes, tempo_preparo, tipo_doce) VALUES (%s, %s, %s, %s)",  # Hannah Reis
                            (receita.nome, ingredientes_str, receita.get_tempo_preparo_hannah_reis(), tipo_doce))  # Hannah Reis
        self.conexao.commit()  # Hannah Reis
        print("Receita inserida com sucesso! ‚úÖ")  # Hannah Reis

    def listar_receitas_hannah_reis(self):  # Hannah Reis
        self.cursor.execute("SELECT * FROM receitas")  # Hannah Reis
        resultados = self.cursor.fetchall()  # Hannah Reis
        for linha in resultados:  # Hannah Reis
            print(f"ID: {linha[0]} | Nome: {linha[1]} | Ingredientes: {linha[2]} | Tempo: {linha[3]} minutos | Tipo de Doce: {linha[4]}")  # Hannah Reis

    def deletar_receita_hannah_reis(self, nome):  # Hannah Reis
        self.cursor.execute("DELETE FROM receitas WHERE nome = %s", (nome,))  # Hannah Reis
        self.conexao.commit()  # Hannah Reis
        print(f"Receita '{nome}' exclu√≠da com sucesso.")  # Hannah Reis

    def fechar_conexao_hannah_reis(self):  # Hannah Reis
        self.conexao.close()  # Hannah Reis

# Teste com banco de dados  # Hannah Reis
bd = BancoDeDadosHannahReis(host="localhost", user="seu_usuario", password="sua_senha", nome_banco="receitas_hannah_reis")  # Hannah Reis

# Criando uma receita e inserindo no banco  # Hannah Reis
bolo = ReceitaHannahReis("Bolo de Laranja", ["laranja", "a√ß√∫car", "ovos", "farinha"], 70)  # Hannah Reis
bolo.exibir_dados_hannah_reis()  # Hannah Reis
bd.inserir_receita_hannah_reis(bolo)  # Hannah Reis

# Criando uma receita da subclasse e inserindo no banco  # Hannah Reis
pudim = ReceitaSobremesaHannahReis(  # Hannah Reis
    "Pudim de Leite",  # Hannah Reis
    ["leite condensado", "leite", "ovos", "a√ß√∫car"],  # Hannah Reis
    90,  # Hannah Reis
    "assado"  # Hannah Reis
)  # Hannah Reis
pudim.exibir_dados_hannah_reis()  # Hannah Reis
bd.inserir_receita_hannah_reis(pudim)  # Hannah Reis

# Listando receitas no banco de dados  # Hannah Reis
bd.listar_receitas_hannah_reis()  # Hannah Reis

# Excluindo receita  # Hannah Reis
bd.deletar_receita_hannah_reis("Bolo de Laranja")  # Hannah Reis

# Fechando conex√£o  # Hannah Reis
bd.fechar_conexao_hannah_reis()  # Hannah Reis

# Teste com banco de dados  # Hannah Reis
bd = BancoDeDadosHannahReis(host="localhost", user="seu_usuario", password="sua_senha", nome_banco="receitas_hannah_reis")  # Hannah Reis

# Criando uma receita e inserindo no banco  # Hannah Reis
minha_receita = ReceitaHannahReis("Escondidinho", ["batata", "carne mo√≠da", "queijo"], 60)  # Hannah Reis
bd.inserir_receita_hannah_reis(minha_receita)  # Hannah Reis

# Listando receitas  # Hannah Reis
bd.listar_receitas_hannah_reis()  # Hannah Reis

# Excluindo receita  # Hannah Reis
bd.deletar_receita_hannah_reis("Escondidinho")  # Hannah Reis

# Fechando conex√£o  # Hannah Reis
bd.fechar_conexao_hannah_reis()  # Hannah Reis
```
